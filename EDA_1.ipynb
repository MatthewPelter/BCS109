The data set being analyzed is the data from all types of cars. What will be examined is the correlation between cylinders, mpg, hp, etc.

---



```
#https://raw.githubusercontent.com/MatthewPelter/BCS109/master/cardata.csv

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import missingno as msno
import seaborn as sns

url = 'https://raw.githubusercontent.com/MatthewPelter/BCS109/master/cardata.csv'
df = pd.read_csv(url)

df.head()
```

```
df.isnull().sum()
```

```
df.nunique()
```

```
df.describe()
```

The following shows the histograms for the different numerical data in the dataset. We can see the amounts of each category like the amount of 4 cylinder vehicles etc.

```
df.hist(figsize=(15,10))
plt.show()
```

With this heatmap, we can see which variables are correlated. It is made obvious that miles per gallon and the amount of cylinders are correlated meaning the more cylinders, the worse mpg. Also, the more horsepower, the less mpg the vehicle has. We can also analyze that the more horsepower the vehicle has, the less time to 60 it has.

```
corrMatrix = df.corr()
sns.heatmap(corrMatrix, annot=True)
plt.show()
```

In this scatterplot, we see the relationship between the horsepower and the time to 60. We see that they have a negative relationship. As the horsepower increases, we see a quicker 0-60.

```
sns.scatterplot(data=df, x='hp', y='time-to-60')
```

This is another great analysis of mpg to cylinder amount. A high engine cylinder count results in a lower mpg and we see that most vehicles have around 20 mpg and most of them are 4 cylinders.

```
with sns.axes_style('white'):
    sns.jointplot(x="mpg", y="cylinders", data=df, kind='kde')
```

This simple count plot just shows how many cars are made by a certain country. We can see that in this dataset, most of the vehicles are made in the US.

```
sns.countplot(x="brand", data=df, palette="Set2")
```
