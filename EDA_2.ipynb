#Heart Failure Prediction
This dataset is used to predict heart failure. Given all the variables such as age, platelets, blood pressure, sex, etc, we can see which variables could potentially cause heart failure.

```
#https://raw.githubusercontent.com/MatthewPelter/BCS109/master/heart_failure_clinical_records_dataset.csv

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import missingno as msno
import seaborn as sns

url = 'https://raw.githubusercontent.com/MatthewPelter/BCS109/master/heart_failure_clinical_records_dataset.csv'
df = pd.read_csv(url)

df.head()
```

```
df.isnull().sum()
```

```
df.nunique()
```

```
df.describe()
```

```
df.hist(figsize=(15,10))
plt.show()
```

This heat map shows which variables are correlated. We can see death and age are strongly correlated. What is interesting is it shows that high blood pressure is strongly correlated with sex. Also, diabetes and age are strongly correlated.

```
df2 = df.drop(['creatinine_phosphokinase', 'ejection_fraction', 'serum_creatinine', 'serum_sodium', 'time'], axis=1)
corrMatrix = df2.corr()
sns.heatmap(corrMatrix, annot=True)
plt.show()
```

This graph just simply shows an average of the ages which we can conclude that the average age is around 60 years old. This was also proved by df.describe()

```
sns.kdeplot(data=df, x='age', fill=True)
```

Distribution between age and amount of platelets in the blood.

###0 = Woman
###1 = Man

```
sns.displot(data=df, x="age", y="platelets", hue="sex")
```

#Conclusion
To conclude the EDA, we see that heart failure can be attributed to high blood pressure, sex, and platelet amount. We also found that high blood pressure is also correlated with sex.
